#### 2.อ่านข้อมูลจากไฟล์ภาพ

```python
import cv2

imgPath = "images/1700022563940.jpg"

img = cv2.imread(imgPath)

#ดูขนาดของภาพ
print(img.shape) 
```

#### 3. Convert color image to grayscale image

```python
gray_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
print(gray_img.shape)
```

#### 4.โหลดตัวจำแนกรูปภาพ ว่ามีอะไรในรูปบ้าง 

```python
face_classifier = cv2.CascadeClassifier(cv2.data.haarcascades+ "haarcascade_frontalface_default.xml")
```

#### 5.face Detact ตรวจจับใบหน้า 

```python
face = face_classifier.detectMultiScale(
    gray_img,
    scaleFactor=1.1,
    minNeighbors=5,
    minSize=(40,40),)
```

##### 6.วาดกรอบสี่เหลี่ยมรอบใบหน้า 

```python
for(x,y,w,h) in face :
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),8) #สีเขียว ความหนา 4 
```

#### 7.Displaying the image แสดงภาพ

```python
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```
